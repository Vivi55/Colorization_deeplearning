```python
from matplotlib import pyplot as plt
import numpy as np
import cv2

# 读入图片
img = cv2.imread('/Research/jupyter/dataset_updated/training_set/lung/covid_ct.jpg')#covid-19lung_02
if img is None:
    print("图片读入失败, 请检查图片路径及文件名")
    exit()


# Matplotlib预设的颜色字符
bgrColor = ('b','g', 'r' )


for cidx, color in enumerate(bgrColor):
    # cidx channel 序号
    # color b / g / r
    cHist = cv2.calcHist([img], [cidx], None, [256], [0, 256])
    # 绘制折线图
    plt.plot(cHist, color=color)  
    #    ll=plt.legend([line1,line2,line3],["Blue", "Green","Red"],loc='upper left')
    plt.ylabel("Amount of pixels",fontsize=14)#设置纵轴单位
    plt.xlabel("Per pixel/bpp",fontsize=14)#设置横轴单位
    plt.title("The rate of RGB in the COVID-19 lung CT image",fontsize=14)
# 设定画布的范围
plt.xlim([0, 260])

# 显示画面
plt.show()
```
